```python
#df_connections = pd.merge(df_connections_start_end,df_connections_end_start,how='inner', left_index=True, right_index=True).stack()
df_count_start_end = df.groupby(['start_station_name','end_station_name']).size().to_frame(name = 'count_connections').reset_index()
df_count_end_start = df.groupby(['end_station_name','start_station_name']).size().to_frame(name = 'count_connections').reset_index()

#df_connections_count['connection_start_end'] = df_connections_count['start_station_name'] + " & " +  df_connections_count['end_station_name']
#df_connections_count['connection_end_start'] = df_connections_count['end_station_name'] + " & " +  df_connections_count['start_station_name']
df_count_start_end['connection'] = df_count_start_end['start_station_name'] + " & " +  df_count_start_end['end_station_name']
df_count_end_start['connection'] = df_count_start_end['end_station_name'] + " & " +  df_count_start_end['start_station_name']

df_connections_count = pd.concat([df_count_start_end, df_count_end_start])

df_connections = pd.DataFrame({'Connections':df_connections_count},index)

df_connections
```
