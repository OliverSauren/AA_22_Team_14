# Boston Interactive Map

The following code is used to show an Interactive Location Map with all the Blubikes stations in Boston. In particular, I used the original Bluebikes data regarding the Bluebikes stations, from which I have eliminated all the stations whose deployment year was higher than 2017. Finally, from this data I obtained an interactive map with all the Blubikes stations in Boston.

```python
import folium
import pandas as pd
```

```python
#import data for markers
data = pd.read_csv("current_bluebikes_stations.csv", header=1)
df = data[data['Deployment Year'] <= 2017]
df
```

```python
#basemap

map = folium.Map(location = [42.3601,-71.0589], zoom_start = 13)

#add markers
for Name, Longitude, Latitude in zip(df['Name'],df['Longitude'],df['Latitude']):
    folium.Marker(location = [Latitude, Longitude], popup = Name).add_to(map)

map
```
