```python
import folium
print(folium.__version__)
import pandas as pd
```

```python
#import data for markers
data = pd.read_excel(r'/Users/leonardogorelli/Downloads/AA_22_Team_14-main/boston_stations.xlsx')
df = pd.DataFrame(data, columns = ["Name","Latitude", "Longitude"])

df
```

```python
#basemap

map = folium.Map(location = [42.3601,-71.0589], zoom_start = 13)

#add markers
for Name, Longitude, Latitude in zip(df['Name'],df['Longitude'],df['Latitude']):
    folium.Marker(location = [Latitude, Longitude], popup = Name).add_to(map)

map
```
