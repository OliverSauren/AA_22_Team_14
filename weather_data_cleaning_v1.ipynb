```python
import pandas as pd
df = pd.read_csv("weather_hourly_boston.csv",  parse_dates=[
                          "date_time"]
    )
df
```

```python
##cut off data from other years, so we only have data for 2017
df = df[(df["date_time"].dt.year > 2016) & (
    df["date_time"].dt.year < 2018)] 
df = df.sort_values("date_time")
df
```

```python
#drop all duplicates
df.drop_duplicates(inplace=True)
df
```

```python
##check for missing values
missing_values = df.isna().sum()
missing_values
```

```python
##fill the missing data with data from the next rows
weatherData_2017 = df.fillna(method='bfill')
```

```python
## save data for 2017
weatherData_2017.to_csv('weather_hourly_boston_2017_cleaned.csv')
```
