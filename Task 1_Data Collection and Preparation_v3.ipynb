# Drop data when necessary

```python
#Delete all raws for which the end time is smaller than or equal to the start time.

df_coreCleaned1 = df_core[df_core.end_time - df_core.start_time >= timedelta(minutes=0)]
number_difference_info(df_core, df_coreCleaned1)
```

```python
#Remove all duplicates (= duplicates of data rows) contained in the dataframe. 
df_coreCleaned2 = df_coreCleaned1.drop_duplicates()
number_difference_info(df_coreCleaned1, df_coreCleaned2)
```

```python
#
df_coreCleaned3 = df_coreCleaned2[np.isfinite(pd.to_numeric(df_coreCleaned2.bike_id, errors="coerce"))]
df_coreCleaned3
number_difference_info(df_coreCleaned2,df_coreCleaned3)
```

```python
#Remove all data rows for which the start or end station is one included by the list of stations to remove from the dataframe.
df_coreCleaned4 = df_coreCleaned3.loc[~(df_coreCleaned3['start_station_name'].isin(list_of_stations_to_remove) | df_coreCleaned3['end_station_name'].isin(list_of_stations_to_remove))]
number_difference_info(df_coreCleaned3, df_coreCleaned4)
```

```python
#Again, group the dataframe by the start station name to get a better readable view. 
#Furthermore, get information on the overall number of entries in the dataframe as well as  
#the number of entries that had been removed from the original dataframe "df_core".

groupByStationNameCleaned = df_coreCleaned4.groupby("start_station_name").describe()
print("The cleaned dataframe now still has "+str(len(df_coreCleaned4))+" entries.")
print("To obtain the cleaned data, the original dataframe with "+str(len(df_core))+" entries was reduced by "+str(len(df_core)-len(df_coreCleaned4))+" rows of data.")
groupByStationNameCleaned
```

# Write cleaned Data

```python
#Set the dataframe "df_coreCleaned4" as the final cleaned dataframe.
df_core_cleaned = df_coreCleaned4
```

```python
#Write the data of the final cleaned dataframe into a csv file named "Boston_2017_Cleaned_Data".
df_core_cleaned.to_csv("Boston_2017_Cleaned_Data.csv")
```
